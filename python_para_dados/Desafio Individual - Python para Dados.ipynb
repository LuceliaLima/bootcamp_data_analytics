**2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mod]**

```python
df_media_moda_mediana = df_new.pivot_table(values='Duração do sono', index='Profissão', aggfunc=[np.mean, pd.Series.mode, np.median])
df_media_moda_mediana
```

```python
media_horas_sono = df_new.groupby('Profissão')['Duração do sono'].mean().round(2)
mediana_horas_sono = df_new.groupby('Profissão')['Duração do sono'].apply(np.median).round(2)
moda_horas_sono = df_new.groupby('Profissão')['Duração do sono'].apply(pd.Series.mode).round(2)

# Imprimir a média das horas de sono para cada profissão
print("Média das horas de sono por profissão:")
print(media_horas_sono)
print()

# Imprimir a mediana das horas de sono para cada profissão
print("Mediana das horas de sono por profissão:")
print(mediana_horas_sono)
print()

# Imprimir a moda das horas de sono para cada profissão
print("Moda das horas de sono por profissão:")
print(moda_horas_sono)
```

**3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?**

```python
df_eng_software = df_new[(df_new['Profissão'] == 'Eng. de Software') & (df_new['Categoria IMC'] == 'Obesidade')]
total_pessoas = len(df_eng_software)

if total_pessoas > 0:
    porcentagem_obesos = (total_pessoas / len(df_new[df_new['Profissão'] == 'Eng. de Software'])) * 100
    print(f"A porcentagem de obesos entre as pessoas que atuam com engenharia de software é: {porcentagem_obesos:.2f}%")
else:
    print("Não há pessoas obesas na categoria de Engenharia de Software no conjunto de dados.")
```

**4. De acordo com os dados, advogar ou ser representante de vendas faz

**5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)**

```python
df_filtered = df_new[df_new['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])]
media_horas_sono = df_filtered.groupby('Profissão')['Duração do sono'].mean().round(2)

profissao_menos_sono = media_horas_sono.idxmin()
menos_horas_sono = media_horas_sono.min()

print(f"Aqueles que são {profissao_menos_sono} têm menos horas de sono,\n"
      f"com uma média de {menos_horas_sono} horas.")
```
