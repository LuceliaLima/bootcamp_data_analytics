**1. Ao visualizar a base voc√™ percebeu que seria melhor alterar o nome de

```python
novos_nomes = {'ID': 'Identificador', 'Press√£o sangu√≠neaaaa': 'Press√£o sangu√≠nea',
               'Ocupa√ß√£o': 'Profiss√£o', 'Categoria BMI': 'Categoria IMC'}
```

**2. Qual √© a m√©dia, a moda e a mediana de horas de sono para cada uma
das profiss√µes? [‚Äòmean‚Äô, np.median, pd.Series.mod]**

```python
df_media_moda_mediana = df_new.pivot_table(values='Dura√ß√£o do sono', index='Profiss√£o', aggfunc=[np.mean, pd.Series.mode, np.median])
df_media_moda_mediana
```

```python
media_horas_sono = df_new.groupby('Profiss√£o')['Dura√ß√£o do sono'].mean().round(2)
mediana_horas_sono = df_new.groupby('Profiss√£o')['Dura√ß√£o do sono'].apply(np.median).round(2)
moda_horas_sono = df_new.groupby('Profiss√£o')['Dura√ß√£o do sono'].apply(pd.Series.mode).round(2)

# Imprimir a m√©dia das horas de sono para cada profiss√£o
print("M√©dia das horas de sono por profiss√£o:")
print(media_horas_sono)
print()

# Imprimir a mediana das horas de sono para cada profiss√£o
print("Mediana das horas de sono por profiss√£o:")
print(mediana_horas_sono)
print()

# Imprimir a moda das horas de sono para cada profiss√£o
print("Moda das horas de sono por profiss√£o:")
print(moda_horas_sono)
```

**3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?**

```python
df_eng_software = df_new[(df_new['Profiss√£o'] == 'Eng. de Software') & (df_new['Categoria IMC'] == 'Obesidade')]
total_pessoas = len(df_eng_software)

if total_pessoas > 0:
    porcentagem_obesos = (total_pessoas / len(df_new[df_new['Profiss√£o'] == 'Eng. de Software'])) * 100
    print(f"A porcentagem de obesos entre as pessoas que atuam com engenharia de software √©: {porcentagem_obesos:.2f}%")
else:
    print("N√£o h√° pessoas obesas na categoria de Engenharia de Software no conjunto de dados.")
```

**4. De acordo com os dados, advogar ou ser representante de vendas faz
voc√™ dormir menos? (Use o m√©todo ‚Äòisin‚Äô, considere a m√©dia)**

**5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o m√©todo ‚Äòisin‚Äô, considere a m√©dia)**

```python
df_filtered = df_new[df_new['Profiss√£o'].isin(['Enfermeiro(a)', 'M√©dico(a)'])]
media_horas_sono = df_filtered.groupby('Profiss√£o')['Dura√ß√£o do sono'].mean().round(2)

profissao_menos_sono = media_horas_sono.idxmin()
menos_horas_sono = media_horas_sono.min()

print(f"Aqueles que s√£o {profissao_menos_sono} t√™m menos horas de sono,\n"
      f"com uma m√©dia de {menos_horas_sono} horas.")
```
